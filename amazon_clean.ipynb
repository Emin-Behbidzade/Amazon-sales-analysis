```python
import pandas as pd
df = pd.read_csv(r'C:\Users\User\Desktop\New folder\Amazon Sale Report.csv', low_memory=False)
```

```python
print(df.shape)        # neçə sətir, neçə sütun
print(df.head())       # ilk 5 sətir
print(df.info())       # sütunların tipi
print(df.isnull().sum()) # boş dəyərlər
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Yoxla
print(df.shape)
print(df.isnull().sum())
print(df['Date'].head())
```

```python
# Ay sütunu əlavə et
df['Month'] = df['Date'].dt.to_period('M')

# 1. Aylıq satış trendi
monthly_sales = df.groupby('Month')['Amount'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)
print(monthly_sales)

# 2. Kateqoriya üzrə satış
category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).reset_index()
print(category_sales)

# 3. Status paylanması
status_dist = df['Status'].value_counts().reset_index()
print(status_dist)

# 4. Ən çox satış olan Top 5 ştat (region)
top_states = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_states)
```

```python
# Təmiz datanı CSV-ə yaz
df.to_csv(r'C:\Users\User\Desktop\New folder\amazon_clean.csv', index=False)

# KPI cədvəllərini də export et
monthly_sales.to_csv(r'C:\Users\User\Desktop\New folder\monthly_sales.csv', index=False)
category_sales.to_csv(r'C:\Users\User\Desktop\New folder\category_sales.csv', index=False)
top_states.to_csv(r'C:\Users\User\Desktop\New folder\top_states.csv', index=False)

print("Fayllar hazırdır!")
```
